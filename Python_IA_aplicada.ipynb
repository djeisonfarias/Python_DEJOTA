#Como criar um chatbot com Python e a IA Gemini

```python
prompt = input("Digite a sua pergunta: ")
print(f"Esta foi a sua pergunta: {prompt}" + ".")

while prompt != "fim":
  resposta = chat.send_message(prompt)
  print(resposta.text)
  print("/n")
  prompt = input("Digite a sua pergunta: ")
  print(f"Esta foi a sua pergunta: {prompt}" + ".")
```

#Resolvendo desafios de programação com o loop while

```python
#loop para incrementar 1.0 para as notas, limitando a nota de ultrapassar 10:

n = 0

while n < 5:
   lista_de_medias[n] = lista_de_medias[n] + 1.0
   if lista_de_medias[n] > 10.0:
       lista_de_medias[n] = 10.0
   n = n + 1

print(lista_de_medias)
```

#Explorando o uso da função len

len = método que apresenta o tamanho da lista, útil quando temos listas com muitos elementos.

```python
n = 0

while n < len(lista_de_medias):
   lista_de_medias[n] = lista_de_medias[n] + 1.0
   if lista_de_medias[n] > 10.0:
       lista_de_medias[n] = 10.0
   n = n + 1

print(lista_de_medias)
```
